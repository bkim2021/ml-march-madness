```python
from random import random
import pandas as pd
import numpy as np
import csv
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('MarchMadness.csv')
t = pd.read_csv('Teams.csv')
n = pd.read_csv('TeamNames.csv')

f = open('MarchMadness.csv', 'r')
lines = f.readlines()
f.close()
```

```python
# The N matrix will initialize the number of games
# The X matrix is the explanatory matrix and displays the schedule/inputs (The amount of team + intercept row)
# The Y matrix contains the responses
N = len(lines)
X = np.zeros((N, len(t)+1))
Y = np.zeros((N, 1))
gameIndex = 0

# The teams dictionary assigns an index to each TeamID
teams = {}
for i in range(len(t)):
    teams[float(t.loc[i, "TeamID"])] = i
    
#TeamID to name
with open('TeamNames.csv', 'r') as infile:
    reader = csv.reader(infile)
    with open('NewTeamNames.csv', 'w') as outfile:
        writer = csv.writer(outfile)
        teamNames = {rows[0]:rows[1] for rows in reader}
```

```python
# Uses gameIndex to assess each game for winners and losers and assigns values accordingly.

for lineIndex in range(0, len(df)):
    homeWon = df.loc[lineIndex, "WLoc"]
    
    winner = df.loc[lineIndex, "WTeamID"]
    loser = df.loc[lineIndex, "LTeamID"]
    
    # Assign a one to the winning team's column and a negative one to the losing team's column.
    
    # If the home team wins, the feature matrix should display a 1.
    # If the home team loses, the feature matrix should display a 0.
    if homeWon == 1:
        X[lineIndex][teams[winner]] = 1
        X[lineIndex][teams[loser]] = -1
        Y[lineIndex] = 1
    else:
        X[lineIndex][teams[winner]] = -1
        X[lineIndex][teams[loser]] = 1
        Y[lineIndex] = 0
    
    # Assign a 1 for the intercept matrix
    X[lineIndex][len(t)] = 1
    lineIndex += 1
    
```

```python
#Setting up components of Newton's method with logistic regression
X = np.matrix(X)
Xt = np.matrix.transpose(X)
XXt = np.dot(X, Xt)
Y = np.matrix(Y)

#Initializing rankings with random values as a starting point
rankings = np.matrix(np.random.rand(len(t)+1, 1))

# Tolerance measures the change between algorithm cycles
tol = 1. 
```

```python
# Newton's method

# 0.00001 works the best
while tol > 0.00001:
    #If tol > 0.00001 the estimated wait time is 15 minutes to run
    p = np.exp(np.dot(X, rankings)) / (1 + np.exp(np.dot(X, rankings)))
    gradient = np.matrix(np.zeros((1, len(t)+1)))
    hessian = 0.4
    
    for i in range(N):
        gradient += np.multiply(X[i], Y[i]) - np.multiply(X[i], p[i])
        hessian += XXt.item((i, i))*p[i]*(1-p[i])
        
    oldrankings = np.matrix.transpose(rankings)
    rankings = oldrankings + (1/hessian)*gradient
    tol = np.dot(rankings-oldrankings, np.matrix.transpose(rankings-oldrankings))
    rankings = np.matrix.transpose(rankings)
    
    print(tol)
```

```python
# Prints out team IDs, team names, and their relative rankings.

print("===============================================")
print("ID\tTeam Name\t\tBradley-Terry")
print("===============================================")

i = 0
for x in teams:
    if len(teamNames[str(int(x))]) < 8:
        text = teamNames[str(int(x))] + "\t\t|\t"
    elif len(teamNames[str(int(x))]) < 16:
        text = teamNames[str(int(x))] + "\t|\t"
    
    print(str(x) + "\t" + text + str(rankings[i]))
    i += 1
```

```python
# Finds the probability

def prob_finder(teamID, oppID):
    intercept = rankings[len(rankings)-1]
    intercept = 0
    
    team = teams[teamID]
    opp = teams[oppID]
    
    teamWeight = rankings[int(team)]
    oppWeight = rankings[int(opp)]
    
    print(teamWeight)
    print(oppWeight)
    
    p = teamWeight - oppWeight + intercept
    
    #print("p is " + str(p))
    prediction = np.exp(p) / (1 + np.exp(p))
    
    return float(prediction[0][0])
```
